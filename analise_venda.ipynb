## 1° Requisito

-Crie um script para simular um dataset de vendas com pelo menos 50 registros

```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
# Lista de produtos
produtos = [
    ("Notebook", "Eletronicos"),
    ("Celular", "Eletronicos"),
    ("Fone de Ouvido", "Eletronicos"),
    ("Camiseta", "Vestuário"),
    ("Tênis", "Vestuário"),
    ("Livro", "Papelaria"),
    ("Caneta", "Papelaria"),
    ("Air Fryer", "Eletrodomésticos"),
    ("Liquidificador", "Eletrodomésticos")
]

print("Produtos carregados:")
for produto, categoria in produtos:
    print(f"- {produto} ({categoria})")
```

```python
# Dicionário de preço fixo
precos_produtos = {
    "Notebook": 2500,
    "Celular": 1200,
    "Fone de Ouvido": 150,
    "Camiseta": 80,
    "Tênis": 250,
    "Livro": 50,
    "Caneta": 5,
    "Air Fryer": 400,
    "Liquidificador": 200
}

print(precos_produtos)
```

```python
# Lista de quantidades
quantidades_possiveis = [1, 2, 3, 5, 10]

# Dias do ano
dias_do_ano = list(range(1, 366))

# Lista vazia
dados = []

#Gerando os registros vendas e colunas
for i in range(1, 51):

    produto, categoria = random.choice(produtos)

    dia = random.choice(dias_do_ano)
    data = datetime(2023, 1, 1) + timedelta(days=dia-1)
    data_formatada = data.strftime('%Y-%m-%d')

    quantidade = random.choice(quantidades_possiveis)

    preco = precos_produtos[produto]

    dados.append([i, data_formatada, produto, categoria, quantidade, preco])

for registro in dados[:50]:
    print(registro)
```

## 2º Requisito

Realize a limpeza dos dados, incluindo:

- Simulação de dados sujos
- Tratamento de valores faltantes
- Remoção de duplicatas
- Conversão de tipos de dados

```python
# Criando tabela dos dados gerados
# Colunas: ID, Data, Produto, Categoria, Quantidade e Preço
vendas = pd.DataFrame(dados, columns=['ID', 'Data', 'Produto', 'Categoria', 'Quantidade', 'Preco'])

print(f"Tabela criada com {len(vendas)} registros")
vendas.head()
```

```python
# Para simulação, será adicionado dados faltantes e duplicados na tabela.

vendas.loc[5, "Preco"] = None
vendas.loc[10, "Quantidade"] = None
vendas.loc[15, "Preco"] = None

linha_duplicada = vendas.iloc[[0]]
vendas = pd.concat([vendas, linha_duplicada], ignore_index=True)

vendas.head(10)
```

```python
# Verificando onde estão os valores nulos

print(vendas.isnull().sum())
vendas[vendas.isnull().any(axis=1)]
```

```python
vendas.to_csv("dados_sujos.csv", index=False)
dados_sujos = pd.read_csv("dados_sujos.csv")
dados_sujos.head(10)
```

# Tratando os valores faltantes, dados duplicados e convertend o tipo de dado.

```python
dados_limpos = pd.read_csv("dados_sujos.csv")

#Removendo linhas duplicadas
antes = len(dados_limpos)
dados_limpos = dados_limpos.copy().drop_duplicates()
depois = len(dados_limpos)


#Utilizando a média para preencher valores nulos
dados_limpos["Preco"] = dados_limpos["Preco"].fillna(dados_limpos["Preco"].mean())
dados_limpos ["Quantidade"] = dados_limpos["Quantidade"].fillna(dados_limpos["Quantidade"].mean())


#Convertendo tipos de dado
dados_limpos["Data"] = pd.to_datetime(dados_limpos["Data"])
dados_limpos["Quantidade"] = dados_limpos["Quantidade"].astype(int)


dados_limpos.head(10)
```

```python
dados_limpos.to_csv("data_clean.csv", index=False)
```

## Analisando os dados limpos para:

- Calcule o total de vendas (Quantidade * Preço) por produto.
- Identifique o produto com o maior número de vendas totais.

```python
# Calculando o total por produto
vendas["Total"] = vendas["Quantidade"] * vendas["Preco"]
total_produto = vendas.groupby("Produto")["Total"].sum()
total_produto = total_produto.sort_values(ascending=False)

print(total_produto)
```

```python
#calculo por produto e maior total em vendas

vendas["Total"] = vendas["Quantidade"] * vendas["Preco"]
total_por_produto = vendas.groupby("Produto")["Total"].sum()
total_por_produto = total_por_produto.sort_values(ascending=False)

print("\nMaior número em vendas totais:")
for produto, valor in total_por_produto.items():
    print(f"{produto}: R$ {valor:.2f}")
```
